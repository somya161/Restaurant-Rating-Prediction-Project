```python
# Dropping unwanted columns
zomato = zomato.drop(['url', 'phone', 'address', 'dish_liked',"location",'menu_item' ,'reviews_list'], axis = 1)
zomato
```

```python
# Renaming the columns
zomato = zomato.rename(columns = {'name' : 'Restaurant'})
zomato = zomato.rename(columns = {'rate' : 'Rating'})
zomato = zomato.rename(columns = {'listed_in(type)' : 'meal_type'})
zomato = zomato.rename(columns = {'listed_in(city)' : 'area'})
zomato = zomato.rename(columns = {'approx_cost(for two people)' : 'approx_cost_for_2'})
```

```python
zomato.head()
```

```python
# Dropping Duplicates
zomato.drop_duplicates(inplace=True)
```

```python
zomato.isnull().sum()
```

### Cleaning Data For Each Column

```python
zomato['Rating'].unique()
```

```python
# Cleaning rate column, removing /5 from data and replacing 'NEW' and '-' with null value
def rating(rate):
    if rate=='-'or rate=='NEW':
        return np.nan
    else:
        rate =str(rate).split('/')[0]
        return float(rate)
    
zomato["Rating"] = zomato["Rating"].apply(rating)
zomato["Rating"].head(11)
```

```python
zomato['approx_cost_for_2'].unique()
```

```python
# Cleaning the cost column
def costing(approx_cost_for_2):
    approx_cost_for_2=str(approx_cost_for_2)
    if "," in approx_cost_for_2:
        approx_cost_for_2 = approx_cost_for_2.replace(",","")
        return float(approx_cost_for_2)
    else:
        return float(approx_cost_for_2) 
        
zomato["approx_cost_for_2"] = zomato["approx_cost_for_2"].apply(costing)
zomato["approx_cost_for_2"].head(11)
```

```python
# Cleaning the restaurant column
zomato['Restaurant'] = zomato['Restaurant'].str.replace('[Ãx][^A-Za-z]+','',regex=True)
zomato
```

```python
sns.heatmap(zomato.isnull())
```

## Dealing with null values

```python
zomato['Rating'] = zomato['Rating'].fillna(zomato['Rating'].mode()[0])
zomato['approx_cost_for_2'] = zomato['approx_cost_for_2'].fillna(zomato['approx_cost_for_2'].mode()[0])
zomato['cuisines'] = zomato['cuisines'].fillna(zomato['cuisines'].mode()[0])
zomato['rest_type'] = zomato['rest_type'].fillna(zomato['rest_type'].mode()[0])
```

### Prepaing data for model training - Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encoding book_table column
encoder_book_table = LabelEncoder()
training_data.book_table = encoder_book_table.fit_transform(training_data.book_table)
```

```python
# Encoding online_order column
encoder_online_order = LabelEncoder()
training_data.online_order = encoder_online_order.fit_transform(training_data.online_order)
```

```python
# Encoding cuisines column
encoder_cuisines = LabelEncoder()
training_data.cuisines = encoder_cuisines.fit_transform(training_data.cuisines)
```

```python
# Encoding rest_type column
encoder_rest_type = LabelEncoder()
training_data.rest_type = encoder_rest_type.fit_transform(training_data.rest_type)
```

```python
# Encoding city column
encoder_area = LabelEncoder()
training_data.area = encoder_area.fit_transform(training_data.area)
```

```python
training_data
```

```python
# We don't need Restaurants column for training purpose so we can drop it
# Dropping restaurant column as the rating is independent of it
training_data.drop(columns=["Restaurant"],inplace=True)
training_data.drop(columns=["meal_type"],inplace=True)
```

```python
# Visualizing heatmap to find coorelation
sns.heatmap(training_data.corr())
```

### Splitting Training and Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
training_data.columns
```

```python
# Independent Variales
features = training_data.drop(columns = ["Rating"])
```

```python
# Dependent Variable
target = training_data["Rating"]
```

```python
# Splitting training and test data
X_train, X_test, y_train, y_test = train_test_split(features,target,test_size=0.1,random_state=10)
```

```python
# finding length of train and test data
#[len(x) for x in [X_train, X_test, y_train, y_test]]
print("Length of X_train is :",[len(x) for x in [X_train]])
print("Length of X_test is :",[len(x) for x in [X_test]])
print("Length of y_train is :",[len(x) for x in [y_train]])
print("Length of t_test is :",[len(x) for x in [y_test]])
```

### 1. Training and Testing with XGBoost Regressor

```python
#!pip install xgboost
```

```python
# Importing the model
from sklearn.metrics import r2_score, mean_squared_error
import xgboost as xg
```

```python
# Defining the model
XG_Model = xg.XGBRegressor(objective ='reg:squarederror',n_estimators = 120, seed = 123)
```

```python
# Fitting the model
XG_Model.fit(X_train, y_train)
```

```python
# Predicting test data using XGBoost
pred = XG_Model.predict(X_test)
```

```python
# Finding r2 score and mean squared error for test data to check model's performance
r2_score_test_xg = r2_score(y_test,pred)
mean_squared_error_test_xg = mean_squared_error(y_test,pred)
print("R_square score of test data is:", r2_score_test_xg)
print("Mean Square Error of test data is :" ,mean_squared_error_test_xg)
```

```python
# Predicting train data using XGBoost
pred_train = XG_Model.predict(X_train)
```

```python
# Finding r2 score and mean squared error for train data to check model's performance
r2_score_train_xg = r2_score(y_train,pred_train)
mean_squared_error_train_xg = mean_squared_error(y_train,pred_train)
print("R_square score of train data is:", r2_score_train_xg)
print("Mean Square Error of train data is :" ,mean_squared_error_train_xg)
```

```python
# Comparing actual and predicted values
df = pd.DataFrame({"y_test" : y_test,"pred":pred})
df
```

```python
# Verifying for the predicted values to be in expected range
(df.pred.max(),df.pred.min())
```
